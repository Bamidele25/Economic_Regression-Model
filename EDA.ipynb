```python
plt.figure(figsize = (7,7))
sns.distplot(mydata['VOA '])
plt.title('VOA Distribution')
```

```python
#scatter plot 
plt.figure(figsize = (7,7))
sns.scatterplot(x= 'GDP ', y = 'VOA ', data = mydata)
plt.title ('Scatter Plot of GDP VS VOA')
```
